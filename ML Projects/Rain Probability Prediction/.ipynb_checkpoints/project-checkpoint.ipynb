```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Temp.csv', delimiter=',',names = ['Avg_temp', 'Humidity', 'Rain'])
data.head()
```

```python
x1 = data.Avg_temp
x2 = data.Humidity
y = data.Rain
```

```python
plt.scatter(x1,y,c='r') 
plt.title('Temperature vs Rain')
plt.xlabel('Temperature')
plt.ylabel('Rain')
plt.show()
```

```python
plt.scatter(x2,y,c='g') 
plt.title('Humidity vs Rain')
plt.xlabel('Humidity')
plt.ylabel('Rain')
plt.show()
```

```python
plt.scatter(x1,x2,c='b') 
plt.title('Temperature vs Humidity')
plt.ylabel('Humidity')
plt.xlabel('Temperature')
plt.show()
```

```python
def u_dis(x1,x2,tp,hm):
    ud={}
    for i in range(len(x1)):
        d = np.sqrt(((x1[i]-tp)**2)+((x2[i]-hm)**2))
        ud[i]=d
    return ud
```

```python
tp = eval(input('Enter Temperature : '))
hm = eval(input('Enter Humidity : '))
```

```python
ud = u_dis(x1,x2,tp,hm)
print(ud)
```

```python
s_ud = sorted(ud.items(), key=lambda kv: kv[1])
print(s_ud)
```

```python
k=3
print("Enter The Value Of K :")
#k=eval(input())
sk=[]
for i in range(k):
    sk.append(s_ud[i][0])
print(sk)
```

```python
ry=[]
for i in range(k):
    ry.append(y[sk[i]])
print(ry)
```

```python
ue,c_ue=np.unique(ry,return_counts=True)
print(ue)
print(c_ue)
```

```python
max_ind=c_ue.argmax(axis=0)
py=ue[max_ind]
print(py)
```

```python
if py == 0:
    print("NO")
else:
    print("YES")
```
