```python
import numpy as np
from sklearn import datasets
```

```python
iris =  datasets.load_iris() 
x = iris.data
y = iris.target
```

```python
print(x[:5])
print(y)
```

```python
def u_dis(x, y):
    ud={}
    for i in range(len(x)):
        d=np.linalg.norm(x[i]-y)
        ud[i]=d
    return ud
```

```python
#xt=[5.1, 3., 4.2, 1.5]
# xt=[6.7, 3.3, 5.7, 2.1]
xt=[]
print("Enter The Value Of TX")
for i in range(len(x[0])):
    t=eval(input())
    xt.append(t)
ud=u_dis(x,xt)
```

```python
print(ud)
```

```python
s_ud = sorted(ud.items(), key=lambda kv: kv[1])
```

```python
print(s_ud)
```

```python
#k=9
print("Enter The Value Of K :")
k=eval(input())
sk=[]
for i in range(k):
    sk.append(s_ud[i][0])
print(sk)
```

```python
ry=[]
for i in range(k):
    ry.append(y[sk[i]])
print(ry)
```

```python
ue,c_ue=np.unique(ry,return_counts=True)
print(ue)
print(c_ue)
```

```python
max_ind=c_ue.argmax(axis=0)
py=ue[max_ind]
print(py)
```
